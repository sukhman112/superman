```python
#1
import random
antal = []
with open("diceRoll.txt", "w") as f1:
    for i in range(1,11):
        tal = random.randint(1,6)
        antal.append(tal)
    f1.write("Simulera 10 tärningskast: " + str(antal))
    newantal = sorted(antal)
    f1.write("\nKastet sorterat: " + str(newantal))
    
    x = len(newantal)
    femmor = []

    for i in range(x):
        if newantal[i] == 5:
            femmor.append(i)
    f1.write("\nAntalet femmor: " + str(len(femmor)))
```

```python
#2
lista =[]
flista = []
elista = []
dlista = []
clista = []
blista = []
alista = []

alla = ["F","E","D","C","B","A"]
with open("Provresultat.txt","r") as f1:
    for rad in f1:
        rad = rad.strip("\n")
        lista.append(rad)


        sak= int(rad[-3:])

        if sak < 20:
            flista.append(rad)

        elif 29 >= sak >= 20:
            elista.append(rad)
            
        elif 39 >= sak >= 30:
            dlista.append(rad)
            
        elif 49 >= sak >= 40:
            clista.append(rad)

        elif 59 >= sak >= 50:
            blista.append(rad)

        elif 70 >= sak >= 60:
            alista.append(rad)
    

    print("\nF")
    for människa in flista:
        print(människa)
    
    print("\nE")
    for människa in elista:
        print(människa)
        
    print("\nD")
    for människa in dlista:
        print(människa)
    
    print("\nC")
    for människa in clista:
        print(människa)
    
    print("\nB")
    for människa in blista:
        print(människa)
    
    print("\nA")
    for människa in alista:
        print(människa)
```

```python
#3
import matplotlib.pyplot as plt
tvåA= []
tvåC = []

with open("NPvt19Ma2A.txt", "r") as f1, open("NPvt19Ma2C.txt", "r") as f2:
    for rad in f1:
        rad = rad.strip("\n")
        rad = rad.strip("%")
        tvåA.append(rad[2:6])

    for rad1 in f2:
        rad1 = rad1.strip("\n")
        rad1 = rad1.strip("%")
        tvåC.append(rad1[2:6])
    
   
    labels = 'A', 'B', 'C', 'D', 'E', 'F'
    sizes = tvåA
    colors = ['gold', 'green', 'blue', 'red', 'orange', 'purple']

    plt.pie(sizes, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.title("NPvt19Ma2A")
    plt.show()

    sizes1 = tvåC
    colors1 = ['gold', 'green', 'blue', 'red', 'orange', 'purple']

    plt.pie(sizes1, labels=labels, colors=colors1,
    autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.title("NPvt19Ma2C")
    plt.show()
```
